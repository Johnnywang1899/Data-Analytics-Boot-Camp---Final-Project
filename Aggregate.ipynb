```python
import glob
import re
import pandas as pd
```

```python
total_data = []
```

```python
folder_path = r'C:\Users\johnn\OneDrive\Desktop\UofT SCS Data Analytics Boot Camp\Module 20 - Final Project\Data-Analytics-Boot-Camp---Final-Project\2000-2021'
all_csv_files = glob.glob(folder_path + "/*.csv")
all_csv_files
```

```python
column_names = ["Date", "Area", "Sales", "Avg_Price", "Med_Price"]
total_data = pd.DataFrame(columns = column_names)
```

```python
for csv_file in all_csv_files:
    df_temp = pd.read_csv(csv_file, sep = ',', parse_dates = ["Date"])
    total_data = pd.concat(objs = [total_data, df_temp], ignore_index = True)
```

```python
total_data.sort_values(by = "Date", ascending = True, inplace = True)
```

```python
total_data = total_data.reset_index(drop = True)
```

```python
total_data.info()
```

```python
total_data[total_data["Sales"] == '0']
```

```python
total_data["Sales"] = total_data["Sales"].astype(str)
```

```python
total_data["Sales"] = total_data["Sales"].apply(lambda x: x.replace(',', ''))
total_data["Sales"] = total_data["Sales"].apply(lambda x: x.replace('-', '0'))
```

```python
total_data["Area"] = total_data["Area"].apply(lambda x: x.replace('-', '0') if int(x[-2:]) < 10 else x.replace('-', ''))
```

```python
total_data["Avg_Price"] = total_data["Avg_Price"].str.replace('-', '0')
total_data["Avg_Price"] = total_data["Avg_Price"].str.replace('$', '')
total_data["Avg_Price"] = total_data["Avg_Price"].str.replace(',', '')

total_data["Med_Price"] = total_data["Med_Price"].str.replace('-', '0')
total_data["Med_Price"] = total_data["Med_Price"].str.replace('$', '')
total_data["Med_Price"] = total_data["Med_Price"].str.replace(',', '')
```

```python
total_data_clean = total_data
```

```python
total_data_clean["Sales"] = total_data_clean["Sales"].astype('int32')
total_data_clean["Avg_Price"] = total_data_clean["Avg_Price"].astype('int32')
total_data_clean["Med_Price"] = total_data_clean["Med_Price"].astype('int32')
```

```python
total_data_clean.info()
```

```python
total_data_clean.to_csv("FINAL.csv", index = False)
```

```python
time_periods = pd.date_range(start = "1999-12-01", end = "2021-04-01", freq = "MS")
```

```python
Sales_output = []
Avg_Price_output = []
for time_period in time_periods:
    df_temp = total_data_clean[total_data_clean["Date"] == time_period]
    Sales_sum = df_temp["Sales"].sum()
    Sales_output.append(Sales_sum)
    df_temp["total_dollar"] = df_temp["Sales"] * df_temp["Avg_Price"]
    total_dollar_sum = df_temp["total_dollar"].sum()
    Avg_Price_month = total_dollar_sum / Sales_sum
    Avg_Price_output.append(Avg_Price_month)
```

```python
final_column_names = ["Date", "Sales_total", "Avg_Price_Month"]
final_df = pd.DataFrame({"Date": time_periods, "Sales_Total": Sales_output, "Avg_Price_Month": Avg_Price_output})
final_df.fillna(0, inplace = True)
final_df.to_csv("FINAL_PRICING.csv", index = False)
```
