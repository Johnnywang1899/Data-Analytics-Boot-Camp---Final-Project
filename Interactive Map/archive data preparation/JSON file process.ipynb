```python
import json
import pandas as pd
import numpy as np
```

```python
with open('forecast.json', mode = 'r') as file:
    result = json.load(file)
```

```python
keys = []
for key in result[0].keys():
    keys.append(key)
```

```python
keys
```

```python
result[0]['2021-05-01']['E02']
```

```python
dates_total = []
keys_total = []
values_total = []
for key in keys:
    keys_temp = []
    for key_temp in result[0][key].keys():
        keys_temp.append(key_temp)
    for my_key in keys_temp:
        dates_total.append(key)
        keys_total.append(my_key)
        values_total.append(result[0][key][my_key])
```

```python
dates_total
```

```python
values_total
```

```python
df = pd.DataFrame({"Date" : dates_total, "Area_Code" : keys_total, "Avg_Price": values_total})
```

```python
df["Avg_Price"] = df["Avg_Price"].astype(float)
```

```python
df
```

```python
df["Total_Condo_Sold_Number"] = np.nan
df["New_Units_Number_construction_complete"] = np.nan
```

```python
df
```

```python
df = df[["Date", "Area_Code", "Total_Condo_Sold_Number", "Avg_Price", "New_Units_Number_construction_complete"]]
```

```python
df.to_csv("for_JSON.csv", index=False)
```
