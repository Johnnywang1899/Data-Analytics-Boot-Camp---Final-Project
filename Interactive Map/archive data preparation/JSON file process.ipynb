```python
import json
import pandas as pd
import numpy as np
```

```python
with open('raw_data.json', mode = 'r') as file:
    result = json.load(file)
```

```python
my_date = []
my_area_code = []
my_total_condo = []
my_avg_price = []
my_new_construction = []
for i in range(0, len(result)):
    index_str = str(i)
    if result[index_str]["Area_Code"] == "All":
        my_date.append(result[index_str]["Date"])
        my_area_code.append(result[index_str]["Area_Code"])
        my_avg_price.append(result[index_str]["Avg_Price"])
        my_total_condo.append(result[index_str]["Total_Condo_Sold_Number"])
        my_new_construction.append(result[index_str]["New_Units_Number_Construction"])
```

```python
my_df = pd.DataFrame({"Date" : my_date, "Area_Code" : my_area_code, "Total_Condo_Sold_Number" : my_total_condo, "Avg_Price": my_avg_price, "New_Units_Number_construction_complete": my_new_construction})
```

```python
my_df.to_csv("All_data.csv", index = False)
```
